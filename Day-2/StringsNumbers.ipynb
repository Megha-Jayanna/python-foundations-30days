# what are Numbers?
Numbers are a fundamental data type in programming that represent numerical values. They can be categorized into different types, such as integers, floating-point numbers, and complex numbers.
Numbers in python supports normal mathematical operations like addition, subtraction, multiplication, division, modulus, exponentiation, and floor division. for example: the plus sign(+) performs addition, the minus sign(-) performs subtraction, the asterisk(*) performs multiplication, the forward slash(/) performs division, the percent sign(%) performs modulus, the double asterisk(**) performs exponentiation, and the double forward slash(//) performs floor division.

Python also provides built-in functions for working with numbers, such as `abs()`, `round()`, `pow()`, and `divmod()`.

```python
print ("addition =", 5 + 3)
print ("subtraction =", 5 - 3)
print ("multiplication =", 5 * 3)
print ("division =", 5 / 3)
print ("modulus =", 5 % 3)
print ("exponentiation =", 5 ** 3)
print ("floor division =", 5 // 3)
```

# What are Strings?
Its used to record both textual and arbitrary collection of data. Strings are enclosed in either single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). They can contain letters, numbers, symbols, and whitespace characters. Strings are immutable, meaning that once they are created, they cannot be modified. However, we can perform various operations on strings, such as concatenation, slicing, and formatting.

**Strings are sequences of one-character strings. They can be indexed and sliced like lists, and they support various methods for manipulation, such as `upper()`, `lower()`, `strip()`, `replace()`, and `split()`. Strings are commonly used for storing and manipulating text data in programming**

# How are Strings and Variables related?
A variable os created when we assign it a value, and the value can be a string. For example, when we write `name = "Guido van Rossum"`, we are creating a variable named `name` and assigning it the string value "Guido van Rossum". The string is created in memory and the variable `name` holds a reference to that string. We can then use the variable `name` to access and manipulate the string as needed.

Strings and variables are closely related because variables are used to store and manipulate string data in programming. Variables provide a way to reference and work with strings, allowing us to perform various operations on them, such as concatenation, slicing, and formatting. By using variables, we can easily manage and manipulate string data in our programs.


```python
# concatenation of strings
first_name = "Guido"
last_name = "van Rossum"
full_name = first_name + " " + last_name
print(full_name)
```

```python
#Slicing of strings
print(name[:5])
print(name[:])
print(name[:4]) # which is same as name[0:4], and it will give us the first 4 characters of the string, which are 'Guid'. The slicing syntax `name[:4]` means "give me the substring of `name` starting from the beginning (index 0) up to, but not including, index 4". So it will return the characters at indices 0, 1, 2, and 3, which are 'G', 'u', 'i', and 'd', respectively.
```

# Immutability of strings
Strings in Python are immutable, which means that once a string is created, it cannot be modified. When we perform operations that seem to modify a string, such as concatenation or slicing, what happens is that a new string is created in memory with the desired modifications, and the original string remains unchanged. For example, when we concatenate two strings, a new string is created that contains the combined value, and the original strings are not altered.



```python
a= "Python"
```

```python
a[0]= 'J' # This will raise a TypeError because strings are immutable and cannot be modified in place. Instead, we can create a new string by concatenating the desired changes:
```

```python
a = 'J' + a[1:] # This will create a new string 'Jython' and assign it back to the variable a.
print(a)
# so python is immutable but we are creating a new string and reassigning it to the variable. so python string will be removed from memory and the new string will be created in memory and assigned to the variable a. this is how we can modify a string in python even though it is immutable.
```

# What are String Literals?
String literals are sequences of characters enclosed in quotes that represent a string value in Python. They can be defined using single quotes (' '), double quotes (" "), or triple quotes (''' ''' or """ """). String literals can contain letters, numbers, symbols, and whitespace characters.

**Note**: Single and double quotes can be used interchangeably for string literals, but triple quotes are typically used for multi-line strings or docstrings. The reason for supporting both single and double quotes is to allow for flexibility in defining strings that may contain one type of quote without needing to escape it. For example, if we want to include a single quote in a string, we can use double quotes to define the string, and vice versa. This helps to avoid the need for escaping characters and makes it easier to work with strings that contain quotes.

```python
# using double quotes to include a single quote in the string
print("I'm learning Python")
# using single quotes to include double quotes in the string
print('She said, "Good!"')

# using single quotes and escaping the single quote with a backslash
print('I\'m learning Python')
# using double quotes and escaping the double quotes with a backslash
print("She said, \"Good!\"")
```

# String formatting
String formatting is a way to create formatted strings by embedding expressions inside string literals. In Python, there are several ways to perform string formatting, including using the `format()` method, f-strings, and the `%` operator.
1. Using the `format()` method:

**Note**:Raw strings are defined by prefixing the string literal with an 'r' or 'R'. They treat backslashes as literal characters and do not interpret them as escape characters. This is particularly useful when dealing with file paths, regular expressions, or any string that contains a lot of backslashes, as it allows us to avoid the need for escaping them. For example, `r"C:\Users\Name"` will be treated as a raw string, and the backslashes will be included in the string without needing to escape them.

```python
#Example using 3 type of string formatting
name = "Guido"
age = 35
# Using the format() method
print("My name is {} and I am {} years old.".format(name, age))

# Using f-strings
print(f"My name is {name} and I am {age} years old.")

# Using the % operator
print("My name is %s and I am %d years old." % (name, age))
```
