# Program-1: Even or odd

```python

# num= int(input("Enter the number: "))
num=12
if(num % 2 ==0):
    print(num, "is an even number")
else:
    print(num, "is an odd number")
```

# Program-2: Palindrome or not

```python

name='madam'
if name == name[::-1]: # The slicing syntax `name[::-1]` is used to reverse the string `name`. It creates a new string that is the reverse of the original string. The `[::-1]` slice notation means "start from the end of the string and move backwards with a step of -1". So, it effectively reverses the order of the characters in the string. By comparing the original string `name` with its reversed version `name[::-1]`, we can determine if the string is a palindrome (i.e., it reads the same forwards and backwards). [start:stop:step]
    print(name, "is a palindrome")
else:
    print(name, "is not a palindrome")
```
