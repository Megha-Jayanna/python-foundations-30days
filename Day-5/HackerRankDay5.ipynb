 # Task
For each input string, print:

all characters at even indices (0, 2, 4, …),
then a space,
then all characters at odd indices (1, 3, 5, …).

```python
# Enter your code here. Read input from STDIN. Print output to STDOUT
indexNum= int(input().strip())
for _ in range(indexNum):
    input_string =input()
    even= input_string[::2]
    odd = input_string[1::2]
    print(even,odd)
```
