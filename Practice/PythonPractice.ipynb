```python
# Comprehensions and Generations

#Example of for loop
res= []
for x in'text':
    # res.append(ord(x))
    res.append(x)
```

```python
res
```

```python
res
```

```python
#Example of map function
res= list(map(ord,'text)'))
res
```

```python
#Example of list comprehension
res= [ord(x) for x in 'text']
res
```

```python
#Example of generator expression
res= (ord(x) for x in 'text')
res
```

```python
#Example in lamda
res= list(map(lambda x: ord(x), 'text'))
res
```
