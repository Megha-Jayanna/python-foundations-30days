# Loops
Loops are used to execute a block of code repeatedly until a certain condition is met. In Python, there are two main types of loops: `for` loops and `while` loops.
1. `for` loops: A `for` loop is used to iterate over a sequence (such as a list, tuple, or string) or other iterable objects.
2.  'while' loops: A `while` loop is used to execute a block of code as long as a specified condition is true.

```python
#syntax for loop
# for variable in sequence:
    # code block to be executed
```

```python
#Sytax for while loop
# while condition:
    # code block to be executed
```

```python
#Example for : for loop
fruits = ['apple', 'banana', 'cherry']
for s in fruits:
    print(s)
```

```python
#Example for while loop
count = 0
while count < 3:
    for i in range(2):
        print(count, i)
        count += 1
```

# Statements in Python
Statements are the building blocks of a Python program. They are instructions that the Python interpreter can execute. There are several types of statements in Python, including:
1. Expression statements: These are statements that evaluate an expression and may have side effects. For example, `x = 5` is an expression statement that assigns the value 5 to the variable x.
2. Compound statements: These are statements that contain other statements. They include `if`, `for`, `while`, `try`, and `with` statements. For example, an `if` statement is a compound statement that contains a block of code that is executed if a specified condition is true.
3. Control flow statements: These are statements that control the flow of execution in a program. They include `break`, `continue`, and `pass` statements. For example, a `break` statement is used to exit a loop prematurely, while a `continue` statement is used to skip the current iteration of a loop and move to the next iteration. A `pass` statement is a placeholder that does nothing and is used when a statement is required syntactically but no action is needed.
4. Function definitions: These are statements that define a function. They start with the `def` keyword followed by the function name and a list of parameters. For example, `def my_function():` is a function definition that defines a function named `my_function` that takes no parameters.
5. Class definitions: These are statements that define a class. They start with the `class` keyword followed by the class name and a colon. For example, `class MyClass:` is a class definition that defines a class named `MyClass`.

```python
#Difference between if and match statement
# The `if` statement is used for conditional branching based on boolean expressions. It allows you to execute different blocks of code based on whether a condition is true or false. For example:
x = 10
if x > 5:
    print("x is greater than 5")
else:
    print("x is not greater than 5")
# The `match` statement, introduced in Python 3.10, is used for pattern matching. It allows you to match a value against a series of patterns and execute code based on which pattern matches. It is more powerful than the `if` statement for certain use cases, such as when you want to match against complex data structures. For example:
def process_value(value):
    match value:
        case 0:
            print("Value is zero")
        case 1:
            print("Value is one")
        case _:
            print("Value is something else")
process_value(0)
process_value(1)
process_value(2)
process_value(_)
```
