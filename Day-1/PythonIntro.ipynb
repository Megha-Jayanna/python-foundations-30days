# How Python runs code?
Python code is executed in three steps:
1. **Compilation**: The Python source code is compiled into bytecode.
    1. The Python interpreter reads the source code and translates it into bytecode.
    2. The bytecode is stored in .pyc files, which are used for faster execution in subsequent runs.
2. **Execution**: The bytecode is executed by the Python virtual machine (PVM).
    1. The PVM reads the bytecode and executes it line by line.
    2. The PVM manages memory, handles exceptions, and performs other tasks necessary for running the program.
3. **Output**: The results of the executed code are produced as output, which can be displayed on the console.

# Key rules for declaring variables
1. Variable names must start with a letter (a-z, A-Z) or an underscore (_).
2. Variable names cannot start with a number (0-9).
3. Variable names can only contain letters, numbers, and underscores.
4. Variable names are case-sensitive (e.g., `myVar` and `myvar` are different).
5. Variable names cannot be the same as Python reserved keywords (e.g., `if`, `for`, `while`, etc.).
6. Variable names should be descriptive and meaningful to improve code readability.
7. Avoid using special characters (e.g., @, $, %, etc.) in variable names.
8. Variable names should not contain spaces; use underscores (_) instead to separate words (e.g., `my_variable`).9Variable names should not be the same as built-in function names (e.g., `print`, `len`, etc.) to avoid confusion and potential errors in the code.
