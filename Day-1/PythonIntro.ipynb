# What is Python?
Python is a interpreted, high-level, general-purpose programming language.
It was created by Guido van Rossum and first released in 1991.
 Python is used for web development, data analysis, AI, scientific computing, and more.


```python
if __name__ == '__main__':
    N = int(input().strip())
    if N % 2 !=0:
        print("Weird")
    elif 2 <= N <= 6:
        print("Not Weird")
    elif 8 <= N <= 20:
        print("Weird but not weird")
    else:
        print(f"entered {N} which is > 20 and even number")
```
